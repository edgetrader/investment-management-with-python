```python
data = erk.get_ind_returns()
data['2013':'2017']
```

```python
quiz_part2 = data['2013':'2017']
```

```python
l = ['Books', 'Steel', 'Oil', 'Mines']
riskfree_rate = 0.1

er = erk.annualize_rets(quiz_part2[l], 12)
cov = quiz_part2.cov().loc[l,l]
```

```python
w_msr = erk.msr(riskfree_rate, er, cov)
w_gmv = erk.gmv(cov)
```
